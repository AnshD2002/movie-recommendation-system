```python
import ast
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
df["genres"] = df["genres"].apply(convert)
df.head()
```

```python
df.dropna(inplace=True)
df
```

```python
df.dropna()
```

```python
df["keywords"] = df["keywords"].apply(convert)
df["keywords"]
```

```python
def convert2(text):
    L=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
df['cast'] = df['cast'].apply(convert2)
df['cast']
```

```python
def convert3(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
        	L.append(i['name'])
    return L
```

```python
df["crew"] = df["crew"].apply(convert3)
df["crew"]
```

```python
df.head()
```

```python
df["overview"][0]
```

```python
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

```python
df.columns
```

```python
df["tags"]=df['genres']+df['keywords']+df['cast']+df['crew']+df['overview']
df["tags"]
```

```python
df["tags"][0]
```

```python
def remove_spaces(obj):
    L=[]
    for i in obj:
        L.append(i.replace(" ",""))
    return L
```

```python
df["tags"] = df["tags"].apply(remove_spaces)
```

```python
drop_columns =['genres', 'keywords', 'cast', 'crew', 'overview']
df=df.drop(drop_columns,axis=1)
```

```python
df
```

```python
df['tags']= df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags']= df['tags'].apply(lambda x:x.lower())
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = CV.fit_transform(df["tags"]).toarray()
vectors
```

```python
vectors.shape
```

```python
CV.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    L=[]
    for i in text.split():
        L.append(ps.stem(i))
    return L
```

```python
df["tags"]=df["tags"].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    dist = sorted(list(enumerate(similarity[movie_index])),reverse = True,key = lambda x:x[1])
    for i in dist[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend("Spectre")
```

```python
import pickle
pickle.dump(similarity,open("similarity.pkl",'wb'))
pickle.dump(df,open("Movies.pkl",'wb'))
```
